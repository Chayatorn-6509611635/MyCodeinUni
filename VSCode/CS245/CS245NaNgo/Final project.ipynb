## 1.Import Libraries

```python
import pandas as pd
import numpy as np
import statistics as st
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2.Read CSV file

```python
df = pd.read_csv('./chip_dataset.csv')
df
```

# Data Preparation

## 1. Scope Table to interest one and clear the one we don't

```python
df_new = df.loc[: , 'Product' : 'Vendor']
df_new
```

## 2.Drop Duplicate

```python
df_new = df_new.drop_duplicates(subset = ['Product'] , keep = 'first')
df_new
```

## 3. Drop NaN of Freq and TDP

```python
df_new.dropna(subset = ['TDP (W)' , 'Freq (MHz)'] , inplace = True)
df_new
```

## 4. Split table By Group CPU and GPU

```python
df_cpu = df_new[df['Type'] == "CPU"]
df_gpu = df_new[df['Type'] == "GPU"]
```

### CPU table :

```python
df_cpu.reset_index(inplace = True , drop = True) # reset index of cpu table
df_cpu
```

### GPU table:

```python
df_gpu.reset_index(inplace = True , drop = True) #reset index of gpu table
df_gpu
```

```python
X = df_cpu['Freq (MHz)']
Y = df_cpu['TDP (W)']
```

```python
plt.scatter(X,Y)
plt.xlabel("Freq (MHz)")
plt.ylabel("TDP (W)")
plt.title("Graph between TDP and Freq (CPU)")
plt.show()
```
